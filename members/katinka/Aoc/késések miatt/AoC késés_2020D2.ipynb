```python
ujlista=[]
for i in lines_list:
    i=list(i.split(" "))
    ujlista.append(i)
ujlista
```

```python
for i in ujlista:
    i[0]=i[0].split("-")
    i[0][0]=int(i[0][0])
    i[0][1]=int(i[0][1])
ujlista
```

```python
ujlista[0][2].count(ujlista[0][1][0])
```

```python
def jojelszavak():
    a=0
    for i in ujlista:
        if i[2].count(i[1][0])>=i[0][0] and i[2].count(i[1][0])<=i[0][1]:
            a=a+1
    print(a)
```

```python
jojelszavak()
```

```python
def jojelszavak2():
    a=0
    for i in ujlista:
        if i[1][0]==i[2][i[0][0]-1]:
            a=a+1
        if i[1][0]==i[2][i[0][1]-1]:
            a=a+1
        if i[1][0]==i[2][i[0][0]-1] and i[1][0]==i[2][i[0][1]-1]:
            a=a-2
    print(a)
```
