```python
with open('input_21_d2.txt', 'r') as fp:
    
    file_contents = fp.read().splitlines()
```

```python
file_contents
```

```python
ujlista=[]
for i in file_contents:
    i=list(i.split(" "))
    ujlista.append(i)
ujlista
```

```python
def melyseg():
    e=0
    l=0
    for i in ujlista:
        if i[0]=='forward':
            e=e+int(i[1])
        elif i[0]=='down':
            l=l+int(i[1])
        elif i[0]=='up':
            l=l-int(i[1])
    return e*l
```

```python
melyseg()
```

```python
def celos():
    aim=0
    elore=0
    depth=0
    for i in ujlista:
        if i[0]=='forward':
            elore=elore+int(i[1])
            depth=depth+aim*int(i[1])
        elif i[0]=='down':
            aim=aim+int(i[1])
        elif i[0]=='up':
            aim=aim-int(i[1])
    return elore*depth
```

```python
celos()
```
